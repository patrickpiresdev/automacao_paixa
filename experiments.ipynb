```python
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib import font_manager as fm

data = [
    {'time': 'vasco', 'doadores': 8, 'apoiadores': 2, 'total presente': 10, 'total pontos': 10},
    {'time': 'flamengo', 'doadores': 5, 'apoiadores': 3, 'total presente': 8, 'total pontos': 8},
    {'time': 'botafogo', 'doadores': 3, 'apoiadores': 2, 'total presente': 5, 'total pontos': 5},
    {'time': 'fluminense', 'doadores': 2, 'apoiadores': 1, 'total presente': 3, 'total pontos': 3},
    {'time': 'palmeiras', 'doadores': 1, 'apoiadores': 1, 'total presente': 2, 'total pontos': 2},
]

fig, ax = plt.subplots(figsize=(8,4))

# set the number of rows and cols for our table

rows = 5
cols = 7

# create a coordinate system based on the number of rows/columns

# adding a bit of padding on bottom (-1), top (1), right (0.5)

ax.set_ylim(1, rows + 2)
ax.set_xlim(1, cols + 1)

font_prop = fm.FontProperties(fname='./assets/fonts/Anton-Regular.ttf')

ax.text(3.5, 6.5, 'Time', va='center', ha='center', fontproperties=font_prop, color='#9d0000')
ax.text(4.5, 6.5, 'Doadores', va='center', ha='center', fontproperties=font_prop, color='#9d0000')
ax.text(5.5, 6.5, 'Apoiadores', va='center', ha='center', fontproperties=font_prop, color='#9d0000')
ax.text(6.5, 6.5, 'Total\nPresente', va='center', ha='center', fontproperties=font_prop, color='#9d0000')
ax.text(7.5, 6.5, 'Total\nPontos', va='center', ha='center', fontproperties=font_prop, color='#9d0000')

rect = patches.Rectangle(
        (1, 6.1),  # bottom left starting position (x,y)
        8,  # width
        0.8,  # height
        ec='none',
        fc='white',
        zorder=-1
    )
ax.add_patch(rect)

for i, row in enumerate(data):
    y = 5.5 - i
    ax.text(1.1, y, f'{i+1}ยบ lugar', va='center', ha='left', fontproperties=font_prop, color='white')
    ax.text(3, y, row['time'].capitalize(), va='center', ha='left', fontproperties=font_prop, color='white')
    ax.text(4.5, y, str(row['doadores']), va='center', ha='center', fontproperties=font_prop, color='white')
    ax.text(5.5, y, str(row['apoiadores']), va='center', ha='center', fontproperties=font_prop, color='white')
    ax.text(6.5, y, str(row['total presente']), va='center', ha='center', fontproperties=font_prop, color='white')
    ax.text(7.5, y, str(row['total pontos']), va='center', ha='center', fontproperties=font_prop, color='white')

    image = mpimg.imread(f"./assets/escudos/{row['time']}.png")
    im = OffsetImage(image, zoom=0.35)
    ab = AnnotationBbox(im, (2.3, y), frameon=False, xycoords="data")
    ax.add_artist(ab)
    
    if i != 0:
        rect = patches.Rectangle(
            (1, 5.1 - i),  # bottom left starting position (x,y)
            7,  # width
            0.8,  # height
            ec='none',
            fc='#c86b7c',
            zorder=-1
        )
        ax.add_patch(rect)
    else:
        rect = patches.Rectangle(
            (1, 5.1 - i),  # bottom left starting position (x,y)
            7,  # width
            0.8,  # height
            ec='none',
            fc='#ffffff',
            zorder=-2
        )
        ax.add_patch(rect)
        rect2 = patches.Rectangle(
            (1.05, 5.15 - i),  # bottom left starting position (x,y)
            6.9,  # width
            0.7,  # height
            ec='none',
            fc='#b5374e',
            zorder=-1
        )
        ax.add_patch(rect2)

# transparent bg
# fig.patch.set_alpha(0)
# ax.patch.set_alpha(0)

# remove axis
ax.axis('off')

plt.savefig('resultado.png', dpi=300, bbox_inches='tight', transparent=True)
```
